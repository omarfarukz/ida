## Matplotlib part 1

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
y
```

```python
x
```

## FUnctional Way

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel('Y Axis')
plt.title("My Test Graph")
plt.show()
```

```python
plt.subplot(1,2,1)#nrow, nclo, figure number/position number
plt.plot(x,y,'g')
plt.xlabel('Row 1')
plt.ylabel('Plot Number 1')
plt.title('Plot Num 1')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
plt.xlabel("Col 1")
plt.ylabel('Row 1')
plt.title('Plot Num 2')
plt.show()
```

## Object Oriented Method

```python
fig= plt.figure()
axes= fig.add_axes([0.1,0.1,0.8,0.8]) #left,Bottom,Weight,Height
axes.plot(x,y)
axes.set_xlabel("X Axis")
axes.set_ylabel('Y Label')
axes.set_title('Object Oriented Graph')
plt.show()
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.6,0.2,0.2,0.2])

axes1.plot(x,y)
axes1.set_title('Learge Plot')
axes2.plot(y,x)
axes.set_title('Smaller Plot')
plt.show()
```

# Part -2 

```python
fig, axes= plt.subplots(figsize= (8,6), nrows=2,ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.show()
```

```python
fig, axes= plt.subplots(figsize= (8,6), nrows=2,ncols=1)
axes[0].plot(x,y)
axes[0].set_xlabel("Xvalues")
axes[0].set_ylabel('Y Label')
axes[1].plot(y,x)
axes[0].set_title("First Plot")
axes[1].set_xlabel('X Label')
axes[1].set_ylabel('Y Label')
axes[1].set_title("Second Plot")
plt.tight_layout()
plt.show()
```

## part 3
** Setting colour, linewidth, linestyle**
## Plot Apperance

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y, color='r')
plt.show()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y, color='r',linewidth=2)
plt.show()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y, color='r',lw=2, alpha = 0.5, ls=':') #line Style
plt.show()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y, color='r',lw=2, alpha = 0.5, ls='--') #line Style
plt.show()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y, color='r',lw=2, alpha = 0.5, ls='-.') #line Style
plt.show()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y, color='r',lw=2, alpha = 0.5, ls='-', marker = '*', markersize=10, markeredgecolor='black') #line Style
plt.show()
```
