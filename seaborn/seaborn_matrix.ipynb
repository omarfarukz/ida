```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
flight = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flight.head()
```

```python
tc = tips.corr(numeric_only=True)
```

```python
fig= plt.figure(figsize=(12,6))
sns.heatmap(data= tc, annot=True)
```

```python
flight
```

```python
fp=flight.pivot_table(index='month',columns='year',values='passengers',observed=False)
```

```python
fp.head()
```

```python
fig = plt.figure(figsize=(12,6))
sns.heatmap(data=fp, annot=True, cmap='coolwarm', linecolor='white',linewidths=1)
plt.tight_layout()
plt.show()
```

```python
sns.clustermap(data=fp)
plt.show()
```

```python
sns.clustermap(data=fp, standard_scale=1)
plt.show()
```

## Seaborn Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.tail()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris, )
plt.show()
```

```python
g = sns.PairGrid(iris,height=8)
g.map(plt.scatter)
plt.show()
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.histplot, kde=True)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot, color='green', fill=True)
plt.show()
```

```python
tips.head()
```

```python
g= sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.histplot,'total_bill','tip')
plt.show()
```

```python
g= sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.scatterplot,'total_bill','tip')
plt.show()
```

## Seaborn Regression Plots

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y ='tip',data=tips, hue='sex')
plt.show()
```

```python
sns.lmplot(x='total_bill',y ='tip',data=tips, hue='sex',markers=['*','v'])
plt.show()
```

```python
sns.lmplot(x='total_bill',y ='tip',data=tips, hue='sex',markers=['*','v'],scatter_kws={'s':100})
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='sex', row='time',)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='day', row='time',hue='sex')
plt.show()
```

## Style and Color

```python
sns.countplot(x=tips['sex'])
sns.despine(left=True,bottom=True)
plt.show()
```
