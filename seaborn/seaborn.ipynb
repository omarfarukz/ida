```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.tail()
```

```python
sns.displot(data=tips, x="total_bill",kde=True, bins=50)
plt.show
```

```python
#Joint Plot
sns.jointplot(x='total_bill', y='tip', data=tips,height=8,kind='reg')
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Tips')
plt.show()
```

```python
#Joint Plot
sns.jointplot(x='total_bill', y='tip', data=tips,height=8,kind='kde',color='skyblue',fill=True)
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Tips')
plt.show()
```

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex', palette='coolwarm')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex', kind='hist')
plt.show()
```

```python
sns.rugplot(tips['total_bill'])
plt.show()
```

```python
sns.histplot(tips['total_bill'])
plt.show()
```

```python
sns.displot(tips['total_bill'])
plt.show()
```

```python
sns.barplot(tips['total_bill'])
plt.show()
```

```python
sns.kdeplot(tips['total_bill'])
plt.show()
```

```python
sns.barplot(data=tips, x='sex', y='total_bill')
plt.show()
```

```python
sns.countplot(tips['sex'])
plt.show()
```

```python
sns.boxenplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips)
plt.show()
```

```python
fig= plt.figure(figsize=(12,6))
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex')
plt.show()
```

```python
fig= plt.figure(figsize=(12,6))
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex',split=True)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
sns.violinplot(x=tips['day'],y=tips["total_bill"])
sns.swarmplot(x=tips["day"],y=tips['total_bill'], color='black')
```
